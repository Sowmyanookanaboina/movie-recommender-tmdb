```python
#original language not necessary as here most of the movies is in english
```

```python
movies=movies[["movie_id","title","overview","genres","keywords","cast","crew"]]
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#have to convert the above type into
# ["Action","Adventre","Fantasy","Scifi"]
```

```python
#the values of genres are string of list of dictionaries
# so we have a special module named ast
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python
movies["genres"]=movies["genres"].apply(convert)
```

```python
movies.head(2)
```

```python
movies["keywords"]=movies["keywords"].apply(convert)
```

```python
movies.head(2)
```

```python
#from cast column we need only top 3 cast hence
def convert3(obj):
    l=[]
    counter =0
    for i in ast.literal_eval(obj):
        if counter !=3:
            l.append(i["name"])
            counter+=1
        else:
            break
    return l
```

```python
movies["cast"]=movies["cast"].apply(convert3)
```

```python
movies.head(2)
```

```python
#we want only director's dictionary to extract..that is we want dictionary with job:director
movies["crew"][0]
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            l.append(i["name"])
            break
    return l
```

```python
movies["crew"]=movies["crew"].apply(fetch_director)
```

```python
movies.head(1)
```

```python
type(movies["overview"][0])
```

```python
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#Sam Worthington needed to be converted into SamWorthington...if keep same as previous they become two different tags that is sam one tag and worthinton one tag
#but in data we have another name Sam Mendes  so the reocmmendation become wrong

movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies["keywords"]=movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"]=movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"]=movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#now create tags columns
movies["tags"]=movies["overview"]+movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]
```

```python
new_df=movies[["movie_id","title","tags"]]
```

```python
new_df
```

```python
new_df["tags"]=new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
#we have an issues here is the words may contain like "action","actions","love","loving","loved" then if we use stemming it becomes ["love","love","love"]
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df["tags"]=new_df["tags"].apply(stem)
```

```python
new_df["tags"][0]
```

```python
new_df["tags"]=new_df["tags"].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
#we have to find the similarity of two movies based on tags of two movies->calculate number of similar words in two tags
#convert tags into vector
#now each movie can become one vector
#the closest vector can be recommended
#text can be converted to vector there are many techiques like bag of vectors
#combine all tags and take 5k most common words and count how many times they appeared in each movie so that it forms a table....i.e in rows it contains movies
#tags and columns contains 5k most common words and values tells how many times they appeared and now the row of one movie is vector for that movie and 
#find 5 closest vectors
#we can't consider stop words->the are for is in a
# we use countvectorizer class
new_df["tags"][0]
```

```python
new_df["tags"][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
# actually it gives cv.fit_transform(new_df["tags"]) a sparse vector so we have to convert it into an array
#each vector is a movie
vectors=cv.fit_transform(new_df["tags"]).toarray()
```

```python
cv.fit_transform(new_df["tags"]).toarray().shape
```

```python
#we have calculate the distance between the vectors we can't calculate euclidian we usine cosine ..
#cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity
```

```python
similarity[0]#it has distance from 1st movie to all other movies,,,,1st value is 1 as the relationship between itself is 1
```

```python
new_df[new_df["title"]=="Avatar"].index[0]#needed to find index of a movie
```

```python
#we want indexes of movies which are closer to the given movie,so we sort them and get first 5 but in sorting process we lose index so to keep index fxed we use enumaret
list(enumerate(similarity[0]))
```

```python
#now sort
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]# now we get 5 closer movies along with indices
```

```python
def recommend(movie):
    movie_index=new_df[new_df["title"]==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend("Avatar")
```
